Read the algorithmic daily returns and clean the data

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
sp500_data_csv = pd.read_csv(sp500_history_csv, parse_dates=True, index_col='Date', infer_datetime_format=True)
sorted_sp500_data_csv = sp500_data_csv.sort_index(ascending=True)
sorted_sp500_data_csv.head(5)
```

```python
# Check Data Types
sorted_sp500_data_csv.dtypes
```

```python
# Fix Data Types
sorted_sp500_data_csv['Close'] = sorted_sp500_data_csv['Close'].str.replace('$', '')
sorted_sp500_data_csv['Close'] = sorted_sp500_data_csv['Close'].astype('float')
sorted_sp500_data_csv.dtypes
```

```python
# Calculate Daily Returns
sorted_sp500_data_csv = sorted_sp500_data_csv.pct_change()
sorted_sp500_data_csv.head(5)
```

```python
# Drop nulls
sorted_sp500_data_csv.dropna(inplace = True)
sorted_sp500_data_csv.head(5)
```

```python
# Rename Column
final_sorted_sp500_data = sorted_sp500_data_csv.rename(columns={"Close" : "S&P 500"})
final_sorted_sp500_data.head(5)
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
combined_returns = pd.concat([whales_data_csv, algo_data_csv, final_sorted_sp500_data], axis="columns", join="inner")
combined_returns.head(5)
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
sharpe_ratios = (combined_returns.mean() * 252) / (combined_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
# Read the first stock
amazon_csv = Path("Resources/AMAZON_DATA.csv")
amazon_data = pd.read_csv(amazon_csv, parse_dates=True, index_col='Date', infer_datetime_format=True) 
amazon_data = amazon_data.sort_index()
amazon_data.head()
```

```python
# Read the second stock
apple_csv = Path("Resources/AP_DATA.csv")
apple_data = pd.read_csv(apple_csv, parse_dates=True, index_col='Date', infer_datetime_format=True) 
apple_data = apple_data.sort_index()
apple_data.head()
```

```python
# Read the third stock
jpm_csv = Path("Resources/JPM_DATA.csv")
jpm_data = pd.read_csv(jpm_csv, parse_dates=True, index_col='Date', infer_datetime_format=True) 
jpm_data = jpm_data.sort_index()
jpm_data.head()
```

```python
# Read the fourth stock
goldman_csv = Path("Resources/GS_DATA.csv")
goldman_data = pd.read_csv(goldman_csv, parse_dates=True, index_col='Date', infer_datetime_format=True) 
goldman_data = goldman_data.sort_index()
goldman_data.head()
```

```python
# Concatenate all stocks into a single DataFrame
aidans_portfolio = pd.concat([amazon_data, apple_data, jpm_data, goldman_data], axis="rows", join="inner")
aidans_portfolio.index = aidans_portfolio.index.normalize()
aidans_portfolio.head()
```

```python
# Reset the index
aidans_portfolio = aidans_portfolio.reset_index()
aidans_portfolio.head()
```

```python
# Pivot so that each column of prices represents a unique symbol
aidans_portfolio = aidans_portfolio.pivot_table(values='Close', index='Date', columns='Symbol')
aidans_portfolio.head()
```

```python
# Drop Nulls
aidans_returns = aidans_portfolio.pct_change()
aidans_returns.dropna(inplace=True)
aidans_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [.25, .25, .25, .25]
aidans_portfolio_returns = aidans_returns.dot(weights)
aidans_portfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
versus_aidan = pd.concat([aidans_portfolio_returns, combined_returns], axis="columns", join="inner")
versus_aidan.columns = ["AIDAN FUND", "SOROS FUND MANAGEMENT LLC", "PAULSON & CO.INC.", "TIGER GLOBAL MANAGEMENT LLC",
                       "BERKSHIRE HATHAWAY INC", "Algo 1", "Algo 2", "S&P 500"]
versus_aidan.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
versus_aidan.dropna(inplace=True)
versus_aidan.head()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others
